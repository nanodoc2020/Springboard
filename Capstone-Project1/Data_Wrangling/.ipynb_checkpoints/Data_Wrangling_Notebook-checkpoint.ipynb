# Capstone Project 1: Data Wrangling Steps

The Seeds data set can be found on the UCI ML Repository. It is simple to download and has a description of the parameters. Upon inspection it is clear that compared to most data sets in use today, this set is very small. There are 210 data points and 8 variables. Seven of these variables are numeric and essentially continuous; they are geometric properties of the seeds taken from precise measurement. The final variable is categorical, and gives the species of the seed with three distinct values: 1="Kama", 2="Rosa", 3="Canadian". The miniscule size is the most challenging aspect of the data, and a number of ways to address this have been considered, including cross-validation, bootstrapping, and creating synthetic data. 

There are no missing values. There is, however, a mismatch of data in correct columns. After converting the original tab-delimited file into CSV format using excel, the mismatched columns were easy to see. Some of the entries had, in random places, been shifted over to the right a column or two, leaving blank spaces where the information should be. All of the data was shifted over uniformly for each occurence, so simply cutting and pasting the rows’ columns into their proper places was all that was required. This misalignment was present in about 10% of the row entries, and took approximately ten minutes of manual copy/paste cleaning to get all of the rows aligned with data in the proper columns.

Certain outliers for each class are seen when the data is visualized in a scatter plot. Clear clusters emerge for each class, but there are about 5 anomalies that breach the class decision boundary and lie in the area for another class. These “outliers” will likely be best dealt with by using hyper-parameter tuning and cross validation. Adding enough nodes in a neural network may help the algorithm classify the outliers properly using other features with which to compare. They are left in the data to provide an accurate sampling of the range of geometric properties for each plant species.

Summary statistics for each non-categorical variable follow.


```python
#Required imports:
import csv
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

#create a dictionary for the species names:
classDict={1:'Kama',2:'Rosa',3:'Canadian'}
classDict
```

```python
#After downloading the data file from the UCI ML repository, we inspect it:
with open('seeds_dataset.txt') as seedstxt:
    print(seedstxt.read())
```

```python
#The tab delimited file is easily imported into excel, saved as a csv file, and viewed:
with open('seeds_data.csv') as seedscsv:
    csv = csv.reader(seedscsv)
    for row in csv:
        print(row)
```

```python
#Next this is cast into a pandas dataframe and inspected:
seeds=pd.read_csv('seeds_data.csv')
print(seeds.head(),'\n\n',seeds.tail())
```

```python
#Now we can use pandas to get some general info about the data:
print(seeds.info(),seeds.describe())
```

## Statistics exploration for non-categorical variables
We examine each variable and look at box plots, histograms, and scatter plots to find relationships.

### Area

```python
seeds.area.describe()
```

```python
#The total average of all species is easy to see, but not very informative:
seeds.area.plot(kind='box')
plt.title('Areas')
plt.ylabel('cm')
```

```python
#The seaborn library easily allows visualization of the areas by class. Clear separation in area
#is seen between classes 2 & 3, while class 1 overlaps both:
sns.boxplot(x='class',y='area',data=seeds)
```

```python
sns.violinplot(x=seeds.area.iloc[0:69])
sns.violinplot(x=seeds.area.iloc[70:139],color='orange')
sns.violinplot(x=seeds.area.iloc[140:209],color='green')
```

```python
seeds.area.iloc[0:69].plot(kind='hist',bins=25)
seeds.area.iloc[70:139].plot(kind='hist',bins=25)
seeds.area.iloc[140:209].plot(kind='hist',bins=25)
```

## Scatter plots

```python
seeds['Name']=seeds['class']
seeds.Name.iloc[0:70]=classDict[1]
seeds.Name.iloc[70:140]=classDict[2]
seeds.Name.iloc[140:210]=classDict[3]

seeds.Name.value_counts()
```

```python
#We can use seaborn to make a scatter plot matrix of all the variables plotted against one another, with the 
#diagonal containing distributions of that variable between each class.
sns.pairplot(seeds,hue='Name')
```

Clusters appear in many of the plots. However, many of them are mixed together making misclassification more likely.
The comparison with the most separated groups appears to be groove length vs. length. Perhaps a clustering algorithm
such as k-nearest neighbors could be useful and informative to explore.

```python
plt.figure(figsize=(14,9))
sns.scatterplot(x='length',y='groove length',data=seeds,hue='Name', size='perimeter', alpha=.9)
```
